3. Write a program in which use all the operators we can use in Python

```python
print(f" {1} + {2} = {1 + 2}")
print(f" {3} - {2} = {3 - 2}")
print(f" {3} * {2} = {3 * 2}")
print(f" {6} / {3} = {6 / 3}")
print(f" {2} ** {3} = {10 ** 3}")

print(f"{2} == {2}: {2 == 2}")
print(f"{2} != {3}: {2 != 3}")
print(f"{3} > {2}: {3 > 2}")
print(f"{2} < {3}: {2 < 3}")
print(f"{4} >= {3}: {4 >= 3}")
print(f"{3} <= {4}: {3 <= 4}")

c = 10
c += 5
print(f"c += 5: {c}")
c -= 3
print(f"c -= 3: {c}")
c *= 2
print(f"c *= 2: {c}")
```
